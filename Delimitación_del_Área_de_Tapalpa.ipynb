```python
# --- FASE 1: CONFIGURACIÓN DEL ENTORNO Y AUTENTICACIÓN DE GEE ---


# 1.2. Importación de librerías
import ee
import folium
import pandas as pd # Para un mejor manejo de los resultados del NDVI medio
import matplotlib.pyplot as plt
import numpy as np


# Autenticación e inicialización de Google Earth Engine
def autenticar_gee(project='ee-nayleths'):
   try:
    ee.Authenticate()
    ee.Initialize(project='ee-nayleths')
    print("Google Earth Engine autenticado e inicializado correctamente.")
   except Exception as e:
    print(f"Error al autenticar/inicializar GEE: {e}")

# 1.4. Función para añadir capas de GEE a un mapa de Folium
def add_ee_layer(self, ee_object, vis_params, name, visible=True):
   try:
        if isinstance(ee_object, ee.image.Image):
            map_id_dict = ee.Image(ee_object).getMapId(vis_params)
            _add_tile_layer(self, map_id_dict, name, visible)

        elif isinstance(ee_object, ee.imagecollection.ImageCollection):
            mosaic = ee_object.mosaic()
            map_id_dict = mosaic.getMapId(vis_params)
            _add_tile_layer(self, map_id_dict, name, visible)

        elif isinstance(ee_object,(ee.geometry.Geometry, ee.feature.Feature, ee.featurecollection.FeatureCollection)):
            geojson = ee_object.getInfo()
            folium.GeoJson(geojson,name=name, style_function=lambda x: {'color': vis_params.get('color', 'black'),'weight': vis_params.get('width', 2)}).add_to(self)

   except Exception as e:
     print("Error al añadir la capa '{name}': {e}")

# Función auxiliar para añadir capas tipo TileLayer
def _add_tile_layer(map_obj, map_id_dict, name, visible):
    folium.TileLayer(
        tiles=map_id_dict['tile_fetcher'].url_format,
        attr='Google Earth Engine',
        name=name,
        overlay=True,
        control=True,
        show=visible
    ).add_to(map_obj)
folium.Map.add_ee_layer = add_ee_layer
# Ejecutar autenticación
autenticar_gee()
```

```python
# --- FASE 3: DATOS SATELITALES Y PREPROCESAMIENTO ---

def cargar_datos_landsat(geometry, start_date='2000-01-01', end_date='2020-12-31'):
    ls5 = (ee.ImageCollection('LANDSAT/LT05/C02/T1_L2')
           .filterBounds(geometry)
           .filterDate(start_date, '2013-05-01')
           .filter(ee.Filter.lt('CLOUD_COVER', 20))
           .select(['SR_B3', 'SR_B4'], ['RED', 'NIR']))

    ls7 = (ee.ImageCollection('LANDSAT/LE07/C02/T1_L2')
           .filterBounds(geometry)
           .filterDate(start_date, end_date)
           .filter(ee.Filter.lt('CLOUD_COVER', 20))
           .select(['SR_B3', 'SR_B4'], ['RED', 'NIR']))

    return ls5.merge(ls7)

# --- FASE 4: CÁLCULO DEL ÍNDICE DE VEGETACIÓN DE DIFERENCIA NORMALIZADA (NDVI) ---

def calcular_ndvi(collection):
    def add_ndvi(image):
        ndvi = image.normalizedDifference(['NIR', 'RED']).rename('NDVI')
        return image.addBands(ndvi).copyProperties(image, ['system:time_start'])
    return collection.map(add_ndvi)


# --- FASE 5: ANÁLISIS TEMPORAL DEL NDVI ---

def calcular_medianas_por_periodo(collection, geometry, periods):
    medianas = {}
    for nombre, (inicio, fin) in periods.items():
        ndvi_filtrado = collection.filterDate(inicio, fin).select('NDVI')
        mediana = ndvi_filtrado.median().clip(geometry)
        medianas[nombre] = mediana
        mapa_tapalpa.add_ee_layer(mediana, ndvi_vis_params, f'NDVI Mediana {nombre}', visible=False)
    return medianas

# --- FASE 6: ESTADÍSTICAS DE NDVI ---

def calcular_estadisticas_ndvi(medianas, geometry, scale=30):
    resultados = {}
    for nombre, imagen in medianas.items():
        media = imagen.reduceRegion(
            reducer=ee.Reducer.mean(),
            geometry=geometry,
            scale=scale,
            maxPixels=1e9
        ).get('NDVI').getInfo()
        resultados[nombre] = media
    return resultados

def extraer_datos_mensuales(collection, geometry, periods, scale=30):
    datos = []
    for nombre, (inicio, fin) in periods.items():
        por_periodo = collection.filterDate(inicio, fin)
        for mes in range(1, 13):
            por_mes = por_periodo.filter(ee.Filter.calendarRange(mes, mes, 'month'))
            media = por_mes.select('NDVI').mean().reduceRegion(
                reducer=ee.Reducer.mean(),
                geometry=geometry,
                scale=scale,
                maxPixels=1e9
            ).get('NDVI')
            desviacion = por_mes.select('NDVI').mean().reduceRegion(
                reducer=ee.Reducer.stdDev(),
                geometry=geometry,
                scale=scale,
                maxPixels=1e9
            ).get('NDVI')

            if media:
                datos.append({
                    'period': nombre,
                    'month': mes,
                    'mean_ndvi': media.getInfo(),
                    'std_dev_ndvi': desviacion.getInfo() if desviacion else 0
                })
    return pd.DataFrame(datos)

def calcular_intervalos_confianza(df, error_multiplier=1):
    df['upper_ci'] = df['mean_ndvi'] + df['std_dev_ndvi'] * error_multiplier
    df['lower_ci'] = df['mean_ndvi'] - df['std_dev_ndvi'] * error_multiplier
    df['lower_ci'] = df['lower_ci'].apply(lambda x: max(x, -1))
    df['upper_ci'] = df['upper_ci'].apply(lambda x: min(x, 1))
    return df

def graficar_ndvi_mensual(df, periods):
    plt.figure(figsize=(12, 7))
    meses = ['Ene', 'Feb', 'Mar', 'Abr', 'May', 'Jun', 'Jul', 'Ago', 'Sep', 'Oct', 'Nov', 'Dic']
    colores =plt.colormaps['viridis']

    for i, (period_name, _) in enumerate(periods.items()):
       period_df = df[df['period'] == period_name].sort_values(by='month')
       color = colores(i / (len(periods) - 1))  # Normalizado

       plt.plot(period_df['month'], period_df['mean_ndvi'], marker='o', linestyle='-',
             label=period_name, color=color)

       plt.errorbar(period_df['month'], period_df['mean_ndvi'],
                 yerr=[period_df['mean_ndvi'] - period_df['lower_ci'],
                       period_df['upper_ci'] - period_df['mean_ndvi']],
                 fmt='none', capsize=3, color=color, alpha=0.6)

    plt.xlabel('Mes')
    plt.ylabel('NDVI Promedio Mensual')
    plt.title('NDVI Mensual con Intervalos de Confianza (2000–2020)')
    plt.xticks(range(1, 13), meses)
    plt.grid(True, linestyle='--', alpha=0.7)
    plt.legend(title='Período')
    plt.tight_layout()
    plt.ylim(0, 1)
    plt.show()

# Parámetros
periods = {
    '2000-2004': ('2000-01-01', '2004-12-31'),
    '2005-2009': ('2005-01-01', '2009-12-31'),
    '2010-2014': ('2010-01-01', '2014-12-31'),
    '2015-2020': ('2015-01-01', '2020-12-31')
}

ndvi_vis_params = {
    'min': 0.1,
    'max': 0.9,
    'palette': ['FFFFFF','CE7E45','DF923D','F1B95B','FCD163','99B718','74A901','66A000','529400','3E8601','004D00']
}

# Flujo
coleccion = cargar_datos_landsat(sierratapalpa)
coleccion_ndvi = calcular_ndvi(coleccion)

median_ndvi_2000_2020 = coleccion_ndvi.select('NDVI').median().clip(sierratapalpa)
mapa_tapalpa.add_ee_layer(
    coleccion_ndvi.select('NDVI').median().clip(sierratapalpa),
    ndvi_vis_params,
    'NDVI Mediana (2000–2020)'
)

medianas_por_periodo = calcular_medianas_por_periodo(coleccion_ndvi, sierratapalpa, periods)
estadisticas = calcular_estadisticas_ndvi(medianas_por_periodo, sierratapalpa)
df_estadisticas = pd.DataFrame(list(estadisticas.items()), columns=['Periodo', 'NDVI Medio'])

df_mensual = extraer_datos_mensuales(coleccion_ndvi, sierratapalpa, periods)
df_mensual = calcular_intervalos_confianza(df_mensual)
graficar_ndvi_mensual(df_mensual, periods)

# --- FASE 7: INTERFAZ DE USUARIO (VISUALIZACIÓN Y CONTROLES EN COLAB) ---
from folium import Map
from folium.plugins import FloatImage
from branca.element import Template, MacroElement

leyenda_html = """
    <details>
    <summary><b>Leyenda del NDVI</b></summary>
    <p>El Índice de Vegetación de Diferencia Normalizada (NDVI) es un indicador de la densidad de la vegetación:</p>
    <ul>
        <li><span style="color:#004D00;">&#9632;</span> <b>0.9–0.7:</b> Vegetación muy densa</li>
        <li><span style="color:#529400;">&#9632;</span> <b>0.7–0.5:</b> Vegetación densa</li>
        <li><span style="color:#74A901;">&#9632;</span> <b>0.5–0.3:</b> Vegetación moderada</li>
        <li><span style="color:#FCD163;">&#9632;</span> <b>0.3–0.1:</b> Vegetación escasa</li>
        <li><span style="color:#DF923D;">&#9632;</span> <b>< 0.1:</b> Agua, nieve, nubes o suelo desnudo</li>
    </ul>
    </details>
    """
# Insertar la leyenda en el mapa
legend = MacroElement()
legend._template = Template(leyenda_html)
mapa_tapalpa.get_root().add_child(legend)


# --- FASE 8: EXPORTACIÓN DE DATOS A GOOGLE DRIVE ---
import os
from google.colab import drive

# Verifica si el punto de montaje existe y contiene archivos
def montar_google_drive():
    if not os.path.isdir('/content/drive') or not os.listdir('/content/drive'):
        drive.mount('/content/drive')

def exportar_shapefile(collection, nombre, carpeta):
    tarea = ee.batch.Export.table.toDrive(
        collection=collection,
        description=f'{nombre}_SHP',
        folder=carpeta,
        fileNamePrefix=f'{nombre.lower()}_boundary',
        fileFormat='SHP'
    )
    tarea.start()
    return tarea

def exportar_geotiff(imagen, nombre, region, carpeta='GEE_Exports', escala=30):
    tarea = ee.batch.Export.image.toDrive(
        image=imagen.float(),
        description=f'{nombre}_GeoTIFF',
        folder=carpeta,
        fileNamePrefix=f'{nombre.lower()}',
        scale=escala,
        region=region,
        maxPixels=1e9
    )
    tarea.start()
    return tarea

# Ejecutar exportaciones
montar_google_drive()
tarea_tapalpa = exportar_shapefile(tapalpa, 'Tapalpa_Municipio_Boundary', 'GEE_Exports')
tarea_ndvi = exportar_geotiff(median_ndvi_2000_2020, 'ndvi_median_2000_2020', sierratapalpa)

# --- FASE 9: ANÁLISIS DE PÉRDIDA DE HÁBITAT FORESTAL (FUTURO) ---
print("\nFase 9: Próximos pasos para el Análisis de Pérdida de Hábitat Forestal (no implementado en este código).")
print("1. Clasificación de Cobertura/Uso del Suelo (usando NDVI y otras bandas).")
print("2. Detección de Cambios (comparación de clasificaciones entre periodos).")
print("3. Análisis Comparativo con ANP (Nevado de Colima).")

print("\n¡Metodología y código completados hasta la exportación de datos!")
```
