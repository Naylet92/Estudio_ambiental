#3. Procesamiento aplicado
El flujo de procesamiento incluyó las siguientes etapas para cada sensor:

3.1. Enmascaramiento de nubes y sombras

Landsat 7 y 8: Se utilizó la banda de calidad QA_PIXEL para excluir píxeles afectados por nubes y sombras.

Sentinel-2: Se utilizó la banda QA60 con las máscaras de nubes cirrus y nubes opacas.

3.2. Corrección topográfica

Se aplicó una corrección por iluminación topográfica utilizando el modelo digital de elevación SRTM (30 m). El método corrige las diferencias de iluminación causadas por la pendiente y orientación del terreno, normalizando las reflectancias con el ángulo solar y el coseno del ángulo de incidencia.

3.3. Composición de mediana

Para cada temporada y sensor se generó un mosaico de mediana utilizando todas las imágenes disponibles. Este enfoque reduce el ruido y minimiza la influencia de artefactos residuales de nubes o sombras.

3.4. Cálculo de índices y componentes

NDVI (Índice de Vegetación de Diferencia Normalizada)---  NDVI = (NIR - Rojo) / (NIR + Rojo)

```python
PREFIJO     = 'XMLA'
NOMBRE      = 'Chamela'
RUTA_AOI    = '/content/drive/MyDrive/ANP/AOI'
RUTA        = f'{RUTA_AOI}/aoi_{PREFIJO}.geojson'
RUTA_GEOJSON= RUTA
RUTA_CSV    = f'{RUTA_AOI}/{PREFIJO}_coordenadas.csv'
CRS_GEO     = 4326
CRS_UTM     = 32613
GEE_PROJECT = 'ee-nayleths'

AÑO             = 2020
COLECCION_L7    = 'LANDSAT/LE07/C02/T1_L2'
COLECCION_L8    = 'LANDSAT/LC08/C02/T1_L2'
COLECCION_S2    = 'COPERNICUS/S2_SR_HARMONIZED'

TEMPORADA_SECA   = ('03-01', '05-31')
TEMPORADA_HUMEDA = ('06-01', '12-31')

# Bandas L7 (igual que L5)
BANDAS_SR       = ['SR_B1', 'SR_B2', 'SR_B3', 'SR_B4', 'SR_B5', 'SR_B7']
BANDAS_CORR     = [b + '_corr' for b in BANDAS_SR]

# Bandas L8 (distinto orden: B2-B7 sin B6)
BANDAS_SR_L8    = ['SR_B2', 'SR_B3', 'SR_B4', 'SR_B5', 'SR_B6', 'SR_B7']
BANDAS_CORR_L8  = [b + '_corr' for b in BANDAS_SR_L8]

# Bandas S2
BANDAS_S2       = ['B2', 'B3', 'B4', 'B8', 'B11', 'B12']

# Coeficientes Tasseled Cap L7 (mismos que L5)
TC_BRIGHT  = [0.3561, 0.3972, 0.3904, 0.6966, 0.2286, 0.1596]
TC_GREEN   = [-0.3344, -0.3544, -0.4556, 0.6966, -0.0242, -0.2630]
TC_WET     = [0.2626, 0.2141, 0.0926, 0.0656, -0.7629, -0.5388]

# Coeficientes Tasseled Cap L8
TC_BRIGHT_L8 = [0.3029, 0.2786, 0.4733, 0.5599, 0.5082, 0.1872]
TC_GREEN_L8  = [-0.2941, -0.2430, 0.5421, 0.2410, -0.5312, -0.3826]
TC_WET_L8    = [0.1511, 0.1973, 0.3283, 0.3407, -0.7117, 0.4559]

# Coeficientes Tasseled Cap S2 (from Shi & Xu, 2019)
TC_BRIGHT_S2 = [0.3037, 0.2793, 0.4743, 0.5585, 0.5082, 0.1863]
TC_GREEN_S2  = [-0.2848, -0.2435, -0.5436, 0.7243, 0.0840, -0.1800]
TC_WET_S2    = [0.1509, 0.1973, 0.3279, 0.3406, -0.7112, -0.4572]

ESCALA_L      = 30
ESCALA_S2     = 10
MAX_NUBES_S2  = 20
ZOOM          = 12
STYLE_AREA    = {'color': 'blue', 'fillColor': '#0000ff30', 'weight': 1.5}
STYLE_BBOX    = {'color': 'red',  'fillColor': '#00000000', 'weight': 2.5}
MAP_TITLE     = 'Área de estudio'
RUTA_IMAGENES = '/content/drive/MyDrive/ANP/Landsat'
```

```python
import ee
import math
import json
import geemap
import geopandas as gpd
from shapely.geometry import box
from google.colab import drive
from IPython.display import display, HTML

# Montar Google Drive
drive.mount('/content/drive')

# Autenticar e inicializar GEE
ee.Authenticate()
ee.Initialize(project=GEE_PROJECT)

print("✓ Entorno listo")
```

```python
# Enmascarar nubes Landsat (L7 y L8)
def mask_clouds(image):
    qa = image.select('QA_PIXEL')
    cloud_shadow_mask = qa.bitwiseAnd(1 << 3).eq(0)
    cloud_mask        = qa.bitwiseAnd(1 << 4).eq(0)
    return image.updateMask(cloud_shadow_mask.And(cloud_mask))

# Enmascarar nubes Sentinel 2
def mask_clouds_s2(img):
    scl  = img.select('SCL')
    mask = scl.neq(3).And(scl.neq(8)).And(scl.neq(9)).And(scl.neq(10))
    return img.updateMask(mask).divide(10000)

#  Corrección topográfica Landsat (L7 y L8)
def correccion_topografica(image, region, bandas_sr, bandas_corr):
    dem        = ee.Image('USGS/SRTMGL1_003').clip(region)
    terreno    = ee.Terrain.products(dem)
    pendiente   = terreno.select('slope').multiply(math.pi / 180)
    orientacion = terreno.select('aspect').multiply(math.pi / 180)
    az_solar   = ee.Number(image.get('SUN_AZIMUTH')).multiply(math.pi / 180)
    elev_solar = ee.Number(image.get('SUN_ELEVATION')).multiply(math.pi / 180)
    cos_i = (
        pendiente.cos().multiply(elev_solar.sin())
        .add(pendiente.sin()
             .multiply(elev_solar.cos())
             .multiply(orientacion.subtract(az_solar).cos()))
    )
    cos_i_safe  = cos_i.max(0.1)
    cos_i_stack = cos_i_safe.rename('cos_i')
    bandas_corr_img = (
        ee.Image.cat([
            image.select(b).multiply(0.0000275).add(-0.2)
                  .divide(cos_i_stack).rename(bandas_corr[i])
            for i, b in enumerate(bandas_sr)
        ])
        .set('SUN_AZIMUTH',   image.get('SUN_AZIMUTH'))
        .set('SUN_ELEVATION', image.get('SUN_ELEVATION'))
        .cast({b: 'double' for b in bandas_corr})
    )
    return bandas_corr_img

# Tasseled Cap
def tasseled_cap(image_corr, bandas_corr, coefs, nombre):
    partes = ee.Image.cat([
        image_corr.select(b).multiply(c)
        for b, c in zip(bandas_corr, coefs)
    ])
    return partes.reduce(ee.Reducer.sum()).rename(nombre)

#  NDVI y TC para L7
def procesar_bandas_l7(mosaico):
    nir  = mosaico.select('SR_B4_corr')
    red  = mosaico.select('SR_B3_corr')
    ndvi = nir.subtract(red).divide(nir.add(red)).rename('NDVI')
    tc_b = tasseled_cap(mosaico, BANDAS_CORR,    TC_BRIGHT,    'TC_Brightness')
    tc_g = tasseled_cap(mosaico, BANDAS_CORR,    TC_GREEN,     'TC_Greenness')
    tc_w = tasseled_cap(mosaico, BANDAS_CORR,    TC_WET,       'TC_Wetness')
    return (ee.Image.cat([ndvi, tc_b, tc_g, tc_w]).clip(region)
              .cast({'NDVI': 'double', 'TC_Brightness': 'double',
                     'TC_Greenness': 'double', 'TC_Wetness': 'double'}))

#  NDVI y TC para L8
def procesar_bandas_l8(mosaico):
    nir  = mosaico.select('SR_B5_corr')
    red  = mosaico.select('SR_B4_corr')
    ndvi = nir.subtract(red).divide(nir.add(red)).rename('NDVI')
    tc_b = tasseled_cap(mosaico, BANDAS_CORR_L8, TC_BRIGHT_L8, 'TC_Brightness')
    tc_g = tasseled_cap(mosaico, BANDAS_CORR_L8, TC_GREEN_L8,  'TC_Greenness')
    tc_w = tasseled_cap(mosaico, BANDAS_CORR_L8, TC_WET_L8,    'TC_Wetness')
    return (ee.Image.cat([ndvi, tc_b, tc_g, tc_w]).clip(region)
              .cast({'NDVI': 'double', 'TC_Brightness': 'double',
                     'TC_Greenness': 'double', 'TC_Wetness': 'double'}))

# NDVI y TC para S2
def procesar_bandas_s2(mosaico):
    nir  = mosaico.select('B8')
    red  = mosaico.select('B4')
    ndvi = nir.subtract(red).divide(nir.add(red)).rename('NDVI')
    tc_b = tasseled_cap(mosaico, BANDAS_S2,      TC_BRIGHT_S2, 'TC_Brightness')
    tc_g = tasseled_cap(mosaico, BANDAS_S2,      TC_GREEN_S2,  'TC_Greenness')
    tc_w = tasseled_cap(mosaico, BANDAS_S2,      TC_WET_S2,    'TC_Wetness')
    return (ee.Image.cat([ndvi, tc_b, tc_g, tc_w]).clip(region)
              .cast({'NDVI': 'double', 'TC_Brightness': 'double',
                     'TC_Greenness': 'double', 'TC_Wetness': 'double'}))
```

```python
gdf = gpd.read_file(RUTA_GEOJSON)
geojson_str = gdf.to_crs(epsg=4326).to_json()
region = ee.FeatureCollection(json.loads(geojson_str)).geometry()

def procesar_temporada(nombre_temporada, fecha_ini, fecha_fin):
    print(f"\n── Temporada: {nombre_temporada} ({fecha_ini} → {fecha_fin}) ──")
    fecha_inicio = f'{AÑO}-{fecha_ini}'
    fecha_fin_   = f'{AÑO}-{fecha_fin}'

    col_l7 = (ee.ImageCollection(COLECCION_L7).filterDate(fecha_inicio, fecha_fin_).filterBounds(region))
    col_l8 = (ee.ImageCollection(COLECCION_L8).filterDate(fecha_inicio, fecha_fin_).filterBounds(region))
    col_s2 = (ee.ImageCollection(COLECCION_S2).filterDate(fecha_inicio, fecha_fin_).filterBounds(region)
                .filter(ee.Filter.lt('CLOUDY_PIXEL_PERCENTAGE', MAX_NUBES_S2)))

    n_l7, n_l8, n_s2 = col_l7.size().getInfo(), col_l8.size().getInfo(), col_s2.size().getInfo()

    print(f"  Imágenes Landsat 7: {n_l7}")
    if n_l7 > 0:
        for f in sorted(col_l7.aggregate_array('DATE_ACQUIRED').getInfo()):
            print(f"    · L7 {f}")
    else:
        print("    Sin imágenes L7")

    print(f"  Imágenes Landsat 8: {n_l8}")
    if n_l8 > 0:
        for f in sorted(col_l8.aggregate_array('DATE_ACQUIRED').getInfo()):
            print(f"    · L8 {f}")
    else:
        print("    Sin imágenes L8")

    print(f"  Imágenes Sentinel 2: {n_s2}")
    if n_s2 > 0:
      fechas_s2 = col_s2.aggregate_array('system:time_start').getInfo()
      fechas_s2 = sorted([ee.Date(f).format('YYYY-MM-dd').getInfo() for f in fechas_s2])
    for f in fechas_s2:
        print(f"    · S2 {f}")
    else:
      print("    Sin imágenes S2")
    #  Procesar L7
    img_l7 = None
    if n_l7 > 0:
        col_l7_corr = (col_l7.map(mask_clouds)
                              .map(lambda img: correccion_topografica(img, region, BANDAS_SR, BANDAS_CORR))
                              .select(BANDAS_CORR))
        img_l7 = procesar_bandas_l7(col_l7_corr.median().clip(region))
        print("  L7 procesado correctamente")
    else:
        print("  L7: Sin datos para esta temporada")

    #  Procesar L8
    img_l8 = None
    if n_l8 > 0:
        col_l8_corr = (col_l8.map(mask_clouds)
                              .map(lambda img: correccion_topografica(img, region, BANDAS_SR_L8, BANDAS_CORR_L8))
                              .select(BANDAS_CORR_L8))
        img_l8 = procesar_bandas_l8(col_l8_corr.median().clip(region))
        print("  L8 procesado correctamente")
    else:
        print("  L8: Sin datos para esta temporada")

    # Procesar S2
    img_s2 = None
    if n_s2 > 0:
        col_s2_limpia = col_s2.map(mask_clouds_s2).select(BANDAS_S2)
        img_s2 = procesar_bandas_s2(col_s2_limpia.median().clip(region))
        print("  S2 procesado correctamente")
    else:
        print("  S2: Sin datos para esta temporada")

    return img_l7, img_l8, img_s2

```

```python

img_seca_l7,   img_seca_l8,   img_seca_s2   = procesar_temporada('Seca',   *TEMPORADA_SECA)
img_humeda_l7, img_humeda_l8, img_humeda_s2 = procesar_temporada('Húmeda', *TEMPORADA_HUMEDA)
def exportar(imagen, sensor, sufijo, escala):
    if imagen is None:
        print(f"Exportación cancelada para '{sensor}_{sufijo}': imagen no disponible.")
        return
    nombre_tarea = f'{PREFIJO}_{AÑO}_{sensor}_{sufijo}'
    tarea = ee.batch.Export.image.toDrive(
        image          = imagen,
        description    = nombre_tarea,
        folder         = 'Chamela_Landsat',
        fileNamePrefix = nombre_tarea,
        region         = region,
        scale          = escala,
        crs            = f'EPSG:{CRS_UTM}',
        maxPixels      = 1e13
    )
    tarea.start()
    print(f"✓ Tarea iniciada: {nombre_tarea}")

exportar(img_seca_l7,   'L7', 'seca',   ESCALA_L)
exportar(img_seca_l8,   'L8', 'seca',   ESCALA_L)
exportar(img_seca_s2,   'S2', 'seca',   ESCALA_S2)
exportar(img_humeda_l7, 'L7', 'humeda', ESCALA_L)
exportar(img_humeda_l8, 'L8', 'humeda', ESCALA_L)
exportar(img_humeda_s2, 'S2', 'humeda', ESCALA_S2)
```
