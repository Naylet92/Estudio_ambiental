## Descripción del área

**1. Caracterización geográfica y fisiográfica**


La reserva se localiza en la vertiente del Pacífico mexicano, en el suroeste del estado de Jalisco, dentro de una región caracterizada por relieves montañosos de origen ígneo y sedimentario. La topografía presenta una combinación de sierras bajas, lomeríos y planicies costeras, con altitudes que oscilan aproximadamente entre 0 y 500 m s.n.m.

Asimismo, la red hidrográfica está compuesta por cuencas pequeñas y arroyos de carácter intermitente, los cuales responden a la marcada estacionalidad de las precipitaciones. Esta configuración influye directamente en la dinámica de escorrentía y en los procesos de erosión del suelo.


**2. Clima**

El clima predominante en la región es cálido subhúmedo con lluvias en verano. Se caracteriza por una temperatura media anual cercana a 24–26 °C y una precipitación media anual que varía entre 700 y 1,200 mm.

En este sentido, la estacionalidad es un rasgo clave: existe una estación seca prolongada (de noviembre a mayo) y una temporada de lluvias concentrada entre junio y octubre. Esta variabilidad climática condiciona fuertemente la estructura y funcionamiento del ecosistema.

**3. Suelos y geología**

Predominan unidades poco desarrolladas, como leptosoles y regosoles, derivados de materiales parentales volcánicos y sedimentarios. Estos suelos suelen ser someros, con baja capacidad de retención de humedad y alta susceptibilidad a la erosión.

Desde el punto de vista geológico, la región presenta una compleja historia tectónica asociada al margen activo del Pacífico, lo que se refleja en la heterogeneidad litológica y en la morfología del terreno.

**4. Cobertura vegetal y uso del suelo**

El paisaje está dominado por selva baja caducifolia, considerada una de las formaciones más representativas del bosque tropical seco en México.

Además, “más del 75% de las especies arbóreas pierden sus hojas durante la estación seca”, lo que constituye una adaptación funcional a la escasez hídrica. También, se identifican parches de selva mediana subcaducifolia, manglares en zonas costeras y vegetación riparia asociada a cauces temporales.

En contraste, en áreas periféricas se observan cambios de uso del suelo vinculados a actividades agropecuarias y desarrollo turístico, lo cual genera procesos de fragmentación del hábitat.

**5. Dinámica ecológica del área**

El área presenta una alta variabilidad interanual en la productividad primaria, estrechamente ligada a la precipitación. En este contexto, estudios ecológicos han demostrado que los bosques tropicales secos “responden rápidamente a las fluctuaciones climáticas, mostrando alta resiliencia pero también vulnerabilidad a disturbios intensos”.

Además, la heterogeneidad espacial del paisaje favorece la coexistencia de múltiples nichos ecológicos, lo que incrementa la biodiversidad y la complejidad estructural del ecosistema.

**6. Relevancia territorial**

El área constituye un sistema socioecológico de alta importancia, ya que integra procesos biofísicos y dinámicas humanas. Su ubicación en la costa del Pacífico la convierte en una zona estratégica para la conservación de corredores biológicos, así como para el análisis de impactos derivados del cambio climático y del cambio de uso de suelo

```python
#  DEFINICIÓN DE VARIABLES

# Prefijo del área
PREFIJO     = 'XMLA'

# Nombre del área
NOMBRE      = 'Chamela'

RUTA_GPKG  = '/content/drive/MyDrive/ANP/ReservadelaBiosferaChamelaCuixmala.gpkg'
LAYER_NAME  = None

# Carpeta de salida en Drive
RUTA_AOI    = '/content/drive/MyDrive/ANP/AOI'

# Rutas derivadas automáticamente (no tocar)
RUTA        = f'{RUTA_AOI}/aoi_{PREFIJO}.geojson'
RUTA_GEOJSON= RUTA
RUTA_CSV    = f'{RUTA_AOI}/{PREFIJO}_coordenadas.csv'

# CRS geográfico y UTM
CRS_GEO     = 4326
CRS_UTM     = 32613

# Proyecto de Google Earth Engine
GEE_PROJECT = 'ee-nayleths'

# Zoom inicial del mapa
ZOOM        = 12

# Estilo del área de estudio
STYLE_AREA  = {'color': 'blue', 'fillColor': '#0000ff30', 'weight': 1.5}

# Estilo del rectángulo rojo (bounding box)
STYLE_BBOX  = {'color': 'red', 'fillColor': '#00000000', 'weight': 2.5}

# Título del mapa
MAP_TITLE   = 'Área de estudio'
```

```python
import ee
import os
import geemap
import pandas as pd
import geopandas as gpd
from shapely.geometry import box
from google.colab import drive
from IPython.display import display, HTML

# Montar Google Drive
drive.mount('/content/drive')

# Autenticar e inicializar GEE
ee.Authenticate()
ee.Initialize(project=GEE_PROJECT)

print("✓ Entorno listo")
```

```python
# Crear carpeta AOI si no existe
os.makedirs(RUTA_AOI, exist_ok=True)

# Cargar y reproyectar a geográfico
gdf = gpd.read_file(RUTA_GPKG, layer=LAYER_NAME).to_crs(epsg=CRS_GEO)
minx, miny, maxx, maxy = gdf.total_bounds

# Reproyectar a UTM
gdf_utm = gdf.to_crs(epsg=CRS_UTM)
minx_u, miny_u, maxx_u, maxy_u = gdf_utm.total_bounds

# Guardar AOI como GeoJSON en Drive
gdf.to_file(RUTA_GEOJSON, driver='GeoJSON')
print(f"AOI guardado en: {RUTA_GEOJSON}")

# Guardar CSV
df = pd.DataFrame([{
    "Area"     : NOMBRE,
    "Lat_min"  : miny,
    "Lat_max"  : maxy,
    "Lon_min"  : minx,
    "Lon_max"  : maxx,
    "UTM_X_min": minx_u,
    "UTM_X_max": maxx_u,
    "UTM_Y_min": miny_u,
    "UTM_Y_max": maxy_u,
    "AOI"      : RUTA_GEOJSON
}])
df.to_csv(RUTA_CSV, index=False)
print(f"CSV guardado en: {RUTA_CSV}")

# Mostrar resultados

html = f"""
<div style="font-family: sans-serif; max-width: 420px;">

  <div style="border: 1px solid #ccc; border-radius: 8px; padding: 16px; margin-bottom: 12px;">
    <b style="font-size: 15px;">Sistema Geodésico: WGS84</b><br><br>
    Latitud: {miny:.6f} – {maxy:.6f}<br>
    Longitud: {minx:.6f} – {maxx:.6f}
  </div>

  <div style="border: 1px solid #ccc; border-radius: 8px; padding: 16px;">
    <b style="font-size: 15px;">Sistema Proyectado: UTM Zona 13N</b><br><br>
    X: {minx_u:.2f} – {maxx_u:.2f}<br>
    Y: {miny_u:.2f} – {maxy_u:.2f}
  </div>

</div>
"""
display(HTML(html))
```

```python
# Cargar el área de estudio y reproyectar a WGS84
gdf = gpd.read_file(RUTA, layer=LAYER_NAME).to_crs(epsg=4326)

# Calcular bounding box
minx, miny, maxx, maxy = gdf.total_bounds
bbox_gdf = gpd.GeoDataFrame(
    geometry=[box(minx, miny, maxx, maxy)], crs=4326
)

# FIX: calcular centroide en CRS proyectado, luego volver a WGS84
centroid_proj = gdf.to_crs(epsg=3857).dissolve().centroid.iloc[0]

centroid = gpd.GeoSeries([centroid_proj], crs=3857).to_crs(epsg=4326).iloc[0]


#  Crear el mapa interactivo
mapa = geemap.Map()
mapa.set_center(centroid.x, centroid.y, ZOOM)

# Agregar área de estudio
mapa.add_gdf(gdf, layer_name=MAP_TITLE, style=STYLE_AREA)

# Agregar rectángulo rojo (bounding box)
mapa.add_gdf(bbox_gdf, layer_name='Bounding box', style=STYLE_BBOX)
print("Representación espacial")
# Mostrar el mapa
mapa
```
