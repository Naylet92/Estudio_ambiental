## Descripción del área

**1. Localización geográfica y contexto territorial**

El área se localiza en el sur del estado de Jalisco, en una zona limítrofe con el estado de Colima, dentro del complejo volcánico del Nevado de Colima. Este sistema montañoso forma parte del Eje Neovolcánico Transversal, caracterizado por su origen volcánico reciente y alta actividad geomorfológica.

Asimismo, el parque se inserta en un gradiente altitudinal pronunciado, con elevaciones que superan los 2,000 m s.n.m. en sus partes más altas, lo que genera una marcada diferenciación de condiciones climáticas y ecológicas a escala local.

**2. Características fisiográficas y geomorfológicas**

El relieve es abrupto y altamente accidentado, dominado por laderas volcánicas, escarpes, barrancas profundas y depósitos piroclásticos. Esta morfología responde a la dinámica eruptiva del Nevado de Colima y del volcán de Fuego de Colima, cuyos procesos han modelado el paisaje actual.

En este sentido, la geodinámica activa condiciona la estabilidad del terreno, así como procesos de erosión, transporte de sedimentos y formación de suelos jóvenes.

**3. Clima**

El clima predomina el tipo templado húmedo a semifrío húmedo, con lluvias en verano y presencia frecuente de neblina, condiciones características del bosque mesófilo de montaña. La temperatura media anual oscila entre 8 y 16 °C, mientras que la precipitación puede superar los 1,500 mm anuales en zonas elevadas.

De manera complementaria, la alta humedad atmosférica y la presencia de nubosidad persistente generan microclimas que favorecen procesos ecológicos particulares, como la captación de agua por intercepción de niebla.

**4. Suelos y geología**

En términos edafológicos, predominan andosoles derivados de cenizas volcánicas, caracterizados por su alta porosidad, buena capacidad de retención de humedad y elevada fertilidad. Sin embargo, estos suelos pueden ser inestables en pendientes pronunciadas.

Desde la perspectiva geológica, la zona está compuesta principalmente por materiales volcánicos recientes (lavas, cenizas y depósitos piroclásticos), lo que influye directamente en la morfogénesis del paisaje.

**5. Cobertura vegetal y uso del suelo**

En relación con la cobertura vegetal, el parque se distingue por la presencia del bosque mesófilo de montaña, considerado uno de los ecosistemas más biodiversos y amenazados de México. Este tipo de vegetación se caracteriza por:

Alta densidad arbórea
Presencia de epífitas (musgos, helechos, orquídeas)
Estratificación compleja
Elevada humedad ambiental

Asimismo, en función del gradiente altitudinal, se identifican transiciones hacia bosques de pino-encino y coníferas en zonas más elevadas.

No obstante, en áreas circundantes se observan presiones por cambio de uso del suelo, principalmente por actividades agrícolas y ganaderas, lo que ha contribuido a la fragmentación del hábitat.

**6. Biodiversidad y dinámica ecológica del área**

Desde una perspectiva ecológica, el área alberga una elevada diversidad de especies, muchas de ellas endémicas o con distribución restringida. El bosque mesófilo de montaña es reconocido por su papel en la conservación de la biodiversidad, ya que “contiene una proporción significativa de especies en un área relativamente pequeña”.

Además, la dinámica ecológica está fuertemente influenciada por factores como la humedad, la altitud y las perturbaciones volcánicas, lo que genera procesos de sucesión ecológica complejos

**7. Relevancia territorial**

El parque desempeña un papel crucial en la regulación hídrica regional, la captura de carbono y la conservación de especies. En consecuencia, su protección resulta estratégica para mantener servicios ecosistémicos esenciales y para mitigar los efectos del cambio climático.

##Definir Variables

```python
#  DEFINICIÓN DE VARIABLES

# Prefijo del área
PREFIJO     = 'NC'

# Nombre del área
NOMBRE      = 'Nevado_Colima'

RUTA  = '/content/drive/MyDrive/ANP/AOI/aoi_nevado.geojson'

# Nombre de la capa dentro del .gpkg
LAYER_NAME  = None

# Carpeta de salida en Drive
RUTA_AOI    = '/content/drive/MyDrive/ANP/AOI'

# CRS geográfico y UTM
CRS_GEO     = 4326
CRS_UTM     = 32613

# Rutas de ficheros de salida usando el prefijo
RUTA_CSV    = f'/content/drive/MyDrive/ANP/AOI/{PREFIJO}_coordenadas.csv'
RUTA_GEOJSON= f'/content/drive/MyDrive/ANP/AOI/{PREFIJO}_aoi.geojson'

# Proyecto de Google Earth Engine
GEE_PROJECT = 'ee-nayleths'

# Zoom inicial del mapa
ZOOM        = 12

# Estilo del área de estudio
STYLE_AREA  = {'color': 'blue', 'fillColor': '#0000ff30', 'weight': 1.5}

# Estilo del rectángulo rojo (bounding box)
STYLE_BBOX  = {'color': 'red', 'fillColor': '#00000000', 'weight': 2.5}

# Título del mapa
MAP_TITLE   = 'Área de estudio'
```

```python
import ee
import os
import geemap
import pandas as pd
import geopandas as gpd
from shapely.geometry import box
from google.colab import drive
from IPython.display import display, HTML

# Montar Google Drive
drive.mount('/content/drive')

# Autenticar e inicializar GEE
ee.Authenticate()
ee.Initialize(project=GEE_PROJECT)

print("✓ Entorno listo")
```

```python
# Crear carpeta AOI si no existe
os.makedirs(RUTA_AOI, exist_ok=True)

# Cargar y reproyectar a geográfico
gdf = gpd.read_file(RUTA, layer=LAYER_NAME).to_crs(epsg=CRS_GEO)
minx, miny, maxx, maxy = gdf.total_bounds

# Reproyectar a UTM
gdf_utm = gdf.to_crs(epsg=CRS_UTM)
minx_u, miny_u, maxx_u, maxy_u = gdf_utm.total_bounds

# Guardar AOI como GeoJSON en Drive
gdf.to_file(RUTA_GEOJSON, driver='GeoJSON')
print(f"AOI guardado en: {RUTA_GEOJSON}")

# Guardar CSV
df = pd.DataFrame([{
    "Area"     : NOMBRE,
    "Lat_min"  : miny,
    "Lat_max"  : maxy,
    "Lon_min"  : minx,
    "Lon_max"  : maxx,
    "UTM_X_min": minx_u,
    "UTM_X_max": maxx_u,
    "UTM_Y_min": miny_u,
    "UTM_Y_max": maxy_u,
    "AOI"      : RUTA_GEOJSON
}])
df.to_csv(RUTA_CSV, index=False)
print(f"CSV guardado en: {RUTA_CSV}")

# Mostrar resultados

html = f"""
<div style="font-family: sans-serif; max-width: 420px;">

  <div style="border: 1px solid #ccc; border-radius: 8px; padding: 16px; margin-bottom: 12px;">
    <b style="font-size: 15px;">Sistema Geodésico: WGS84</b><br><br>
    Latitud: {miny:.6f} – {maxy:.6f}<br>
    Longitud: {minx:.6f} – {maxx:.6f}
  </div>

  <div style="border: 1px solid #ccc; border-radius: 8px; padding: 16px;">
    <b style="font-size: 15px;">Sistema Proyectado: UTM Zona 13N</b><br><br>
    X: {minx_u:.2f} – {maxx_u:.2f}<br>
    Y: {miny_u:.2f} – {maxy_u:.2f}
  </div>

</div>
"""
display(HTML(html))
```

```python
# Cargar el área de estudio y reproyectar a WGS84
gdf = gpd.read_file(RUTA, layer=LAYER_NAME).to_crs(epsg=4326)

# Calcular bounding box
minx, miny, maxx, maxy = gdf.total_bounds
bbox_gdf = gpd.GeoDataFrame(
    geometry=[box(minx, miny, maxx, maxy)], crs=4326
)

# FIX: calcular centroide en CRS proyectado, luego volver a WGS84
centroid_proj = gdf.to_crs(epsg=3857).dissolve().centroid.iloc[0]

centroid = gpd.GeoSeries([centroid_proj], crs=3857).to_crs(epsg=4326).iloc[0]


#  Crear el mapa interactivo
mapa = geemap.Map()
mapa.set_center(centroid.x, centroid.y, ZOOM)

# Agregar área de estudio
mapa.add_gdf(gdf, layer_name=MAP_TITLE, style=STYLE_AREA)

# Agregar rectángulo rojo (bounding box)
mapa.add_gdf(bbox_gdf, layer_name='Bounding box', style=STYLE_BBOX)
print("Representación espacial")
# Mostrar el mapa
mapa
```
