## Descripción del área

**1. Localización geográfica y contexto territorial**
En primer lugar, el polígono se localiza claramente en la región centro-occidental de Jalisco, abarcando principalmente territorios de los municipios de:

Mixtlán
Atengo
Atenguillo (zona marginal)
Proximidad a Cuautla

Asimismo, el área se ubica al noroeste inmediato de la Sierra de Quila, funcionando como una zona periférica estratégica, lo cual es clave en términos de conectividad ecológica.

**2. Características fisiográficas y geomorfológicas**

El área se inserta en la transición entre el Eje Neovolcánico Transversal y sistemas serranos locales del occidente.

Presenta:

Relieve montañoso elongado
Altitudes aproximadas entre 1,200 y 2,300 m s.n.m.
Pendientes fuertes, especialmente en flancos orientales

Geomorfológicamente predominan:

Mesetas volcánicas disectadas
Lomeríos estructurales
Barrancas asociadas a drenaje dendrítico

**3. Clima**

El clima dominante es:

Templado subhúmedo con lluvias en verano (Cw)

Con variaciones locales:

Semicálido subhúmedo en zonas más bajas

Parámetros característicos:

Temperatura media anual: 16–22 °C
Precipitación: 700–1,100 mm/año

**4. Suelos y geología**

La base geológica está dominada por:

Rocas volcánicas (andesitas, basaltos)
Depósitos piroclásticos

Respecto a los suelos:

Andosoles (alta capacidad de retención hídrica)
Luvisoles (fertilidad media-alta)
Cambisoles

**5. Cobertura vegetal y uso del suelo**

La cobertura vegetal muestra un patrón de mosaico:

Bosque de pino-encino en zonas altas
Encinares secundarios
Pastizales inducidos
Áreas agrícolas (principalmente maíz de temporal)

Asimismo, el uso del suelo está marcado por:

Ganadería extensiva
Fragmentación del bosque
Expansión agrícola

**6. Biodiversidad y dinámica ecológica**

La región presenta características de zona de transición ecológica, lo que incrementa su diversidad:

Fauna: venado cola blanca (Odocoileus virginianus), puma (Puma concolor)
Avifauna diversa de bosque templado
Flora con especies de pino (Pinus spp.) y encino (Quercus spp.)

En términos dinámicos:

Fragmentación del hábitat
Procesos de regeneración secundaria
Reducción de conectividad estructural

**7. Relevancia ambiental**

Esta área adquiere alta relevancia por:

Su función como zona de amortiguamiento indirecta de la Sierra de Quila
Su papel en la conectividad del paisaje forestal
Su contribución a la regulación hídrica regional
Su potencial para restauración ecológica

Además, representa un caso ideal para evaluar procesos de cambio de uso de suelo en zonas no protegidas, lo cual es consistente con estudios que señalan que “las matrices agroforestales determinan la viabilidad ecológica de las ANP”

##Definir Variables

```python
#  DEFINICIÓN DE VARIABLES

# Prefijo del área
PREFIJO     = 'Verde'

# Nombre del área
NOMBRE      = 'Sierra_Verde'

RUTA  = '/content/drive/MyDrive/ANP/AOI/aoi_sierra verde.geojson'

# Nombre de la capa dentro del .gpkg
LAYER_NAME  = None

# Carpeta de salida en Drive
RUTA_AOI    = '/content/drive/MyDrive/ANP/AOI'

# CRS geográfico y UTM
CRS_GEO     = 4326
CRS_UTM     = 32613

# Rutas de ficheros de salida usando el prefijo
RUTA_CSV    = f'/content/drive/MyDrive/ANP/AOI/{PREFIJO}_coordenadas.csv'
RUTA_GEOJSON= f'/content/drive/MyDrive/ANP/AOI/{PREFIJO}_aoi.geojson'

# Proyecto de Google Earth Engine
GEE_PROJECT = 'ee-nayleths'

# Zoom inicial del mapa
ZOOM        = 12

# Estilo del área de estudio
STYLE_AREA  = {'color': 'blue', 'fillColor': '#0000ff30', 'weight': 1.5}

# Estilo del rectángulo rojo (bounding box)
STYLE_BBOX  = {'color': 'red', 'fillColor': '#00000000', 'weight': 2.5}

# Título del mapa
MAP_TITLE   = 'Área de estudio'
```

```python

import ee
import os
import geemap
import pandas as pd
import geopandas as gpd
from shapely.geometry import box
from google.colab import drive
from IPython.display import display, HTML

# Montar Google Drive
drive.mount('/content/drive')

# Autenticar e inicializar GEE
ee.Authenticate()
ee.Initialize(project=GEE_PROJECT)

print("✓ Entorno listo")
```

```python
# Crear carpeta AOI si no existe
os.makedirs(RUTA_AOI, exist_ok=True)

# Cargar y reproyectar a geográfico
gdf = gpd.read_file(RUTA, layer=LAYER_NAME).to_crs(epsg=CRS_GEO)
minx, miny, maxx, maxy = gdf.total_bounds

# Reproyectar a UTM
gdf_utm = gdf.to_crs(epsg=CRS_UTM)
minx_u, miny_u, maxx_u, maxy_u = gdf_utm.total_bounds

# Guardar AOI como GeoJSON en Drive
gdf.to_file(RUTA_GEOJSON, driver='GeoJSON')
print(f"AOI guardado en: {RUTA_GEOJSON}")

# Guardar CSV
df = pd.DataFrame([{
    "Area"     : NOMBRE,
    "Lat_min"  : miny,
    "Lat_max"  : maxy,
    "Lon_min"  : minx,
    "Lon_max"  : maxx,
    "UTM_X_min": minx_u,
    "UTM_X_max": maxx_u,
    "UTM_Y_min": miny_u,
    "UTM_Y_max": maxy_u,
    "AOI"      : RUTA_GEOJSON
}])
df.to_csv(RUTA_CSV, index=False)
print(f"CSV guardado en: {RUTA_CSV}")

# Mostrar resultados

html = f"""
<div style="font-family: sans-serif; max-width: 420px;">

  <div style="border: 1px solid #ccc; border-radius: 8px; padding: 16px; margin-bottom: 12px;">
    <b style="font-size: 15px;">Sistema Geodésico: WGS84</b><br><br>
    Latitud: {miny:.6f} – {maxy:.6f}<br>
    Longitud: {minx:.6f} – {maxx:.6f}
  </div>

  <div style="border: 1px solid #ccc; border-radius: 8px; padding: 16px;">
    <b style="font-size: 15px;">Sistema Proyectado: UTM Zona 13N</b><br><br>
    X: {minx_u:.2f} – {maxx_u:.2f}<br>
    Y: {miny_u:.2f} – {maxy_u:.2f}
  </div>

</div>
"""
display(HTML(html))
```

```python
# Cargar el área de estudio y reproyectar a WGS84
gdf = gpd.read_file(RUTA, layer=LAYER_NAME).to_crs(epsg=4326)

# Calcular bounding box
minx, miny, maxx, maxy = gdf.total_bounds
bbox_gdf = gpd.GeoDataFrame(
    geometry=[box(minx, miny, maxx, maxy)], crs=4326
)

# FIX: calcular centroide en CRS proyectado, luego volver a WGS84
centroid_proj = gdf.to_crs(epsg=3857).dissolve().centroid.iloc[0]

centroid = gpd.GeoSeries([centroid_proj], crs=3857).to_crs(epsg=4326).iloc[0]


#  Crear el mapa interactivo
mapa = geemap.Map()
mapa.set_center(centroid.x, centroid.y, ZOOM)

# Agregar área de estudio
mapa.add_gdf(gdf, layer_name=MAP_TITLE, style=STYLE_AREA)

# Agregar rectángulo rojo (bounding box)
mapa.add_gdf(bbox_gdf, layer_name='Bounding box', style=STYLE_BBOX)
print("Representación espacial")
# Mostrar el mapa
mapa
```
